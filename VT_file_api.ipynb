```python
import requests
import sys
import re
import json
import pandas as pd
import datetime
import os
```

```python
VIRUSTOTAL_APIKEY = os.getenv('VIRUSTOTAL_APIKEY')
request_headers = {
    'x-apikey': VIRUSTOTAL_APIKEY.strip(),
}
```

```python
with open('/Users/eloui/Downloads/hashes2.txt') as f:
    hash_list = f.readlines()
    
for line in hash_list:
    line = line.strip()

hash_df = pd.DataFrame()
```

```python
for i in hash_list:
    try:
        url = "https://www.virustotal.com/api/v3/files/" + i
        r = requests.get(url, headers=request_headers)
        data = r.json()
        
        df = pd.json_normalize(data)
        hash_df = pd.concat([hash_df, df])
    except Exception as e:
        print(e)
```
