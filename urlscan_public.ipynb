```python
import requests
import pandas as pd
import os
import urllib.parse
pd.set_option('display.max_columns', 100)
```

```python
URLSCAN_API_KEY = os.getenv('URLSCAN_API_KEY')

request_headers = {
    'api-key': URLSCAN_API_KEY,
}
```

```python
output = pd.DataFrame()
```

```python
querylist = []
with open("urlscanquery.txt", "r") as input:
    for line in input:
        line = line.strip()
        querylist.append(line)
```

### Step 1
Query Search API to find matches for our query, list results in DF

```python
for q in querylist:
    try:
        query = urllib.parse.quote(q, safe='') # this is necessary to encode the query parameters correctly in the URL

        url = "https://www.urlscan.io/api/v1/search?q=" + query
        r = requests.get(url, headers=request_headers)
        data = r.json()
        results = data["results"]

        results_df = pd.json_normalize(results)             
        output = pd.concat([output, results_df])

    except Exception as e: print(e)
```

```python
output
```

### Step 2
Create DataFrame for Result API data.requests dictionary, which will create one row for each request in a given result

```python
total_df = pd.DataFrame()
```

```python
for u in output['_id']:
    try:
        query = urllib.parse.quote(u, safe='')
        url = "https://www.urlscan.io/api/v1/result/" + query
        r = requests.get(url, headers=request_headers)
        data = r.json()
        df = pd.json_normalize(data['data']['requests'])
        total_df = pd.concat([total_df, df])
    except Exception as e: print(e)
```

```python
df = pd.json_normalize(data['data']['requests'])
```

#### Subset the DataFrame to look for rows (i.e., requests) whose requests.request.url parameter contains the specified string, and list the response.hash value - dropping duplicates and null/NaN values

```python
hashes_list = total_df[total_df['request.request.url'].str.contains('cdncloud\.services', na=False)]['response.hash'].drop_duplicates().dropna().values
```

```python
print(hashes_list)
```

### Step 3

Query the /responses/ endpoint to retrieve and download the raw content of each file, here formjacking scripts

```python
for h in hashes_list:
    try:
        query = urllib.parse.quote(h, safe='')
        url = "https://urlscan.io/responses/" + query
        r = requests.get(url, headers=request_headers)
        content = r.content
        with open(h, "wb") as file: #have to use the b flag because the content is of type 'bytes'
            file.write(content)
        
    except Exception as e: print(e)
```
