```python
import pandas as pd
from passivetotal import analyzer
analyzer.init()
analyzer.set_date_range(start='2021-01-01', end='2022-08-15') 
```

```python
%%bash
echo "hivnd.com" > output.txt
```

```python
with open('output.txt', encoding='utf-8') as foo:
    f = foo.readlines()

output = pd.DataFrame()

for item in f:
    host = analyzer.Hostname(item.strip())
    parents = host.hostpair_parents
    parents_df = parents.as_df
    output = pd.concat([output, parents_df])
```
